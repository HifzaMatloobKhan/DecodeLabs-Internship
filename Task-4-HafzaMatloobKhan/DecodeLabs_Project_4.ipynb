```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ds = pd.read_excel("Cleaned_DecodeLabs.xlsx")

print(ds.head())
```

```python
top_products = ds["Product"].value_counts().head(10)

top_products.plot(kind="bar", figsize=(10,5))

plt.title("Wireless Earbuds generated the highest order volume")
plt.xlabel("Products")
plt.ylabel("Orders")

plt.xticks(rotation=45)

plt.savefig("bar_chart.png")

plt.show()
```

```python
ds["Month"] = ds["Date"].dt.month

monthly_sales = ds.groupby("Month")["TotalPrice"].sum()

plt.plot(
    monthly_sales.index,
    monthly_sales.values,
    marker="o"
)

plt.title("Revenue increased during Q4 sales periods")
plt.xlabel("Month")
plt.ylabel("Revenue")

plt.savefig("line_chart.png")

plt.show()
```

```python
sns.scatterplot(
    x=ds["Quantity"],
    y=ds["TotalPrice"]
)

plt.title("Higher quantity orders generate higher revenue")
plt.xlabel("Quantity")
plt.ylabel("Total Price")

plt.savefig("scatter_plot.png")

plt.show()
```

```python
pivot = pd.pivot_table(
    ds,
    values="TotalPrice",
    index="Product",
    columns="PaymentMethod",
    aggfunc="sum"
)

pivot.head(5).plot(
    kind="bar",
    stacked=True,
    figsize=(10,5)
)

plt.title("Digital payments generate most product revenue")
plt.xlabel("Products")
plt.ylabel("Revenue")

plt.savefig("stacked_bar.png")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.scatterplot(
    x=ds["Quantity"],
    y=ds["TotalPrice"]
)

plt.title(
    "Higher order quantities strongly correlate with increased transaction revenue",
    fontsize=11,
    fontweight="bold"
)

plt.xlabel("Quantity")
plt.ylabel("Total Price")

plt.savefig("scatter_plot.png")

plt.show()
```

```python
data = ds[[
    "Quantity",
    "UnitPrice",
    "ItemsInCart",
    "TotalPrice"
]]

sns.heatmap(
    data.corr(),
    annot=True,
    cmap="coolwarm"
)

plt.title("Transaction value depends on quantity and pricing")

plt.savefig("heatmap.png")

plt.show()
```

```python
monthly_sales.plot(kind="bar")

plt.ylim(50000, 60000)

plt.title("Misleading Chart with Truncated Axis")

plt.show()
```

```python
top_products.plot(kind="bar")

plt.title("Few products dominate customer demand")

plt.xticks(rotation=45)

plt.show()
```

```python
top_products.plot(
    kind="bar",
    color="red"
)

plt.grid(True)

plt.title("Messy Visualization")

plt.show()
```

```python
sns.histplot(
    ds["TotalPrice"],
    kde=True
)

plt.title("Distribution of Total Price")

plt.savefig("distribution_plot.png")

plt.show()
```
