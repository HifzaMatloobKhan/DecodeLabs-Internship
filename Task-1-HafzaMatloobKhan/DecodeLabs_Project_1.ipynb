## Data Preprocessing
This tutorial uses standard python libraries used for data processing and cleaning


```python
import pandas as pd

ds = pd.read_excel("DecodeLabs.xlsx")

print(ds.head())
```

```python
print(ds.info())

print(ds.describe())
```

```python
ds["CouponCode"] = ds["CouponCode"].fillna("No Coupon")
```

```python
print(ds.isnull().sum())
```

```python
duplicates = ds.duplicated().sum()

print("Duplicate Rows:", duplicates)
```

```python
duplicate_ids = ds["OrderID"].duplicated().sum()

print("Duplicate Order IDs:", duplicate_ids)
```

```python
ds["Date"] = pd.to_datetime(ds["Date"], errors="coerce")
```

```python
invalid_dates = ds["Date"].isnull().sum()

print("Invalid Dates:", invalid_dates)
```

```python
text_columns = [
    "Product",
    "PaymentMethod",
    "OrderStatus",
    "ReferralSource"
]

for col in text_columns:
    ds[col] = ds[col].str.strip()
```

```python
numeric_columns = ["Quantity", "UnitPrice", "ItemsInCart", "TotalPrice"]

for col in numeric_columns:
    ds[col] = pd.to_numeric(ds[col], errors="coerce")
```

```python
print(ds[numeric_columns].dtypes)
```

```python
calculated_total = ds["Quantity"] * ds["UnitPrice"]

incorrect_totals = (
    round(calculated_total, 2) != round(ds["TotalPrice"], 2)
).sum()

print("Incorrect TotalPrice Rows:", incorrect_totals)
```

```python
print("\nFINAL VALIDATION")
print("----------------")

print("\nMissing Values:")
print(ds.isnull().sum())

print("\nDuplicate Rows:")
print(ds.duplicated().sum())

print("\nDuplicate Order IDs:")
print(ds["OrderID"].duplicated().sum())

print("\nInvalid Dates:")
print(ds["Date"].isnull().sum())
```

```python
ds.to_excel("Cleaned_DecodeLabs.xlsx", index=False)

print("\nCleaned dataset saved successfully!")
```
